```python
import pandas as pd

# Read the CSV with error handling
df = pd.read_csv('datasource/items.csv', encoding='utf-8')
# or try:
# df = pd.read_csv('your_file.csv', encoding='latin1')

# Clean the data and save a new CSV
df_clean = df.copy()

# Remove problematic characters from text columns
text_columns = ['Inactive', 'Item', 'Variety', 'Color', 'LabelDescription', 'Definition', 'PictureLayout', 'PictureLink', 'SunConditions']

for col in text_columns:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].astype(str).str.encode('ascii', errors='ignore').str.decode('ascii')

# Save cleaned CSV
df_clean.to_csv('datasource/items_cleaned.csv', index=False, encoding='utf-8')
```

```python
# Create mapping for TypeID = 0 items
type_mapping = {
    'Mum': 7,                    # Perennial
    'Thunbergia': 6,            # Annual
    'Clematis': 7,              # Perennial
    'Ple': 7,                   # Assuming perennial
    'Hosta': 7,                 # Perennial
    'Pumpkin': 8,               # Vegetable
    'Salvia': 6,                # Annual (most common type)
    'Cucumber-Pickle': 8,        # Vegetable
    'Aster': 7,                 # Perennial
    'Geranium': 6,              # Annual
    'Begonia': 6,               # Annual
    'Disporum': 7,              # Perennial
    'Mesculn-Pak choi': 8,      # Vegetable
    'Campanula': 7              # Perennial
}

# Apply the mapping
for item_name, new_type_id in type_mapping.items():
    items_clean.loc[(items_clean['TypeID'] == 0) & (items_clean['Item'] == item_name), 'TypeID'] = new_type_id

# Check if any TypeID = 0 items remain
remaining = items_clean[items_clean['TypeID'] == 0]['Item'].unique()
if len(remaining) > 0:
    print(f"Items still with TypeID = 0: {remaining}")

items_clean
```

```python
items_clean.to_csv('items_categorized.csv', index=False)
print("Categorization complete!")
```

```python
valid_type_ids = [3, 4, 6, 7, 8, 11, 12, 13]

print("Invalid TypeIDs found:", items_clean[~items_clean['TypeID'].isin(valid_type_ids)]['TypeID'].unique())

# Map any invalid TypeIDs to a default (like 7 for Perennial)
items_clean.loc[~items_clean['TypeID'].isin(valid_type_ids), 'TypeID'] = 7

print("After fixing - TypeID counts:")
print(items_clean['TypeID'].value_counts().sort_index())

items_clean.to_csv('forced_valid_typeids.csv', index=False)
```
